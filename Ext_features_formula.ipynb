```python
import pandas as pd
import numpy as np
from data_reader import *
from features_helper import FeaturesHelper
pd.options.display.max_columns = None
pd.options.display.max_rows = None

df = read_all_csvs()
helper = FeaturesHelper(df)
all_features = helper.get_available_features()
helper.run_features_list(all_features, log=False)
```

---

## Rolling price changes standard deviation 
`15m/60m_chg_std`

This feature is simply computed by taking the standard deviation of the 1-minute price movements of a fixed time-interval. In our case, we have `15m` and `60m`.

```python
helper.plot_feature(['15m_chg_std', '60m_chg_std'], n=500)
```

---

## 1-minute Price Change
`1m_price_chg`

This feature is simply a change (in bps) in price per minute.

```python
helper.plot_feature(['1m_price_chg'], n=500)
```

---

## 15/60-minute Price Change
`15/60m_price_chg`

Same as above with 15-minute and 60-minute interval.

```python
helper.plot_feature(['15m_price_chg', '60m_price_chg'], n=500)
```

---

## Rolling draw up from peak
`15m/60m_draw_down`

This feature is computed by taking the maximum price reached in a past fixed time interval, and calculating the change from it to the current close price.

```python
helper.plot_feature(['15m_draw_down', '60m_draw_down'], n=500)
```

---

## Rolling draw up from trough
`15m/60m_draw_up`

Similar to `draw_down` This feature is computed by taking the minimum price attained in a past fixed time interval, and calculating the change from it to the current close price.

```python
helper.plot_feature(['15m_draw_up', '60m_draw_up'], n=500)
```

---

## Rolling standardised price
`15m/60m_z_price`

Giving a fixed time interval, this feature is computed by taking a moving average and moving standard deviation of the price, and calculating the standard score of the last price of that rolling period.

```python
helper.plot_feature(['15m_z_price', '60m_z_price'], n=500)
```

---

## Rolling standardised volume
`15m/60m_z_volume`

Giving a fixed time interval, this feature is computed by taking a moving average and moving standard deviation of the volume traded, and calculating the standard score of the last volume of that rolling period.

```python
helper.plot_feature(['15m_z_volume', '60m_z_volume'], n=500)
```

---

## Average Directional Index
`adx`

Detailed formula is available at [StockCharts](<https://school.stockcharts.com/doku.php?id=technical_indicators:average_directional_index_adx>).

This feature is a popular technical indicator to measure both direction and strength of price trends.

```python
helper.plot_feature(['adx'], n=500)
```

---

## Fast/Slow Stochastic Oscillator
`fso`/`sso`

Detailed formula is available at [Investopedia](<https://www.investopedia.com/terms/s/stochasticoscillator.asp>).

These two features are momentum indicators generating oversold/overbought signals.

```python
helper.plot_feature(['fso', 'sso'], n=500)
```

---

## Moving Average Convergence Divergence
`MACD`

Detailed formula is available at [Investopedia](<https://www.investopedia.com/terms/m/macd.asp>)

As mentioned before, MACD is a trend-following momentum indicator to help find out certain price up/down trends.

```python
helper.plot_feature(['macd'], n=500)
```

---

## On-Balance Volume
`obv`

Detailed formula is available at [Investopedia](<https://www.investopedia.com/terms/o/onbalancevolume.asp>)

This features try to make use of volume flow by +/- sign to predict future stock price movement.

But it is subject to extreme values problem. As you can see from the graph below, the mean is very likely not reverting to 0.

```python
helper.plot_feature(['obv'], n=500)
```

---

## Relative Strength Index
`RSI`

Detailed formula is available at [Investopedia](<https://www.investopedia.com/terms/r/rsi.asp>)

This features measure the magnitude of price movement to predict oversold/overbought signals.The number is subtracted by 50 for 0-mean purpose.

```python
helper.plot_feature(['rsi'], n=500)
```

---

## Volume Changes
`volume_chg`/`5m_smoothed_volume_chg`

This features simply computes the minute-on-minute change in volume traded.

Smoothed volume change means to do a 5-minute moving average of the volume changes.

```python
helper.plot_feature(['volume_chg', '5m_smoothed_volume_chg'], n=500)
```

---

## Log Volume
`log_volume`
